```python
from numpy import array
from numpy import mean
from numpy import cov
from numpy.linalg import eig
A=array([[1,2],[3,4],[5,6]])
print("----------ARRAY---------")
print(A)
```

```python
M=mean(A.T,axis=1)
print("----------MEAN---------")
print(M)
```

```python
C=A-M
print("----------*---------")
print(C)
```

```python
V=cov(C.T)
print("----------*---------")
print(V)
```

```python
values,vectors=eig(V)
print("----------*---------")
print(vectors)
print("----------*---------")
print(values)
```

```python
P = vectors.T.dot(C.T)
print(P.T)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data=pd.read_csv("https://raw.githubusercontent.com/Rishi56584/sample/refs/heads/main/50_startups.csv")
data.head()
```

```python
x=data.iloc[:,0:4]
y=data.iloc[:,4]
x.head()
```

```python
y.head()
```

```python
states=pd.get_dummies(x['State'],drop_first=True)
x=x.drop('State',axis=1)
x=pd.concat([x,states],axis=1)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=42)
from sklearn.linear_model import LinearRegression
regression=LinearRegression()
regression.fit(x_train,y_train)
```

```python

y_pred=regression.predict(x_test)
from sklearn.metrics import r2_score
score=r2_score(y_test,y_pred)
score
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
import warnings
warnings.filterwarnings("ignore")
data=load_iris()
x=data.data
y=data.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
rf_cif=RandomForestClassifier(n_estimators=100,random_state=42)
rf_cif.fit(x_train,y_train)
```

```python
y_pred=rf_cif.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(f"Accuary of Random Forest Classifiers {accuracy*100:2f}%")
```

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.datasets import load_iris
import warnings
warnings.filterwarnings("ignore")
data=load_iris()
x=data.data
y=data.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=42)
base_estimator=DecisionTreeClassifier(max_depth=1)
ada_df=AdaBoostClassifier(estimator=base_estimator,n_estimators=50,learning_rate=10,random_state=42)
#ada_df=AdaBoostClassifier(base_estimator=base_estimator,n_estimators=50,learning_rate=10,random_state=42)

ada_df.fit(x_train,y_train)
y_pred=ada_df.predict(x_test)
accuracy=accuracy_score(y_test,y_pred)
print(f"Accuary of Adaboost Classifiers {accuracy*100:2f}%")
```

```python

from sklearn.cluster import Birch
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from numpy import random
random.seed(1)
x,_=make_blobs(n_samples=400,centers=5,cluster_std=1.2)
plt.scatter(x[:,0],x[:,1])
plt.title("Generated Clustering Data")
plt.show()
bclust=Birch(branching_factor=200,
              threshold=1).fit(x)
print("BIRCH parameters:",bclust.get_params())
labels=bclust.predict(x)
plt.scatter(x[:,0],x[:,1],c=labels)
plt.title("BIRCH Clustering")
plt.show()
```
